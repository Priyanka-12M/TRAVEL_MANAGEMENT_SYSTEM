```python
import pandas as pd
from sqlalchemy import create_engine

# Create the SQLAlchemy engine with the correct format
engine = create_engine(f'mysql+mysqlconnector://{user}:{password}@{host}:{port}/{database}')

# Read the tables into DataFrames
users = pd.read_sql('SELECT * FROM users', con=engine)
flights = pd.read_sql('SELECT * FROM flights', con=engine)
hotels = pd.read_sql('SELECT * FROM hotels', con=engine)
payments = pd.read_sql('SELECT * FROM payments', con=engine)
bookings=pd.read_sql('SELECT * FROM bookings', con=engine)


# Printing the first 5 records from each table
print("users Data:")
print(users.head())

print("\nflights Data:")
print(flights.head())

print("\nhotels Data:")
print(hotels.head())

print("\npayments Data:")
print(payments.head())

print("\nbookings Data:")
print(bookings.head())
```

# **Exploratory Data Analysis (EDA) and Visualization:**
# **Grouping users by city and counting the number of users in each city** **bold text**

```python
# Grouping usernames by address and counting the number of users in each city
username = users.groupby('address')['user_id'].count().reset_index()

# Renaming the columns for clarity
username.columns = ['address', 'Total users']

# Displaying the city-wise user count
print("User Name Count:")
print(username)
```

# **Plotting a bar chart for city-wise total number of users**

```python
import matplotlib.pyplot as plt

# Grouping users by city and counting the number of unique users in each city
username = users.groupby('address')['user_id'].nunique().reset_index()

# Plotting a bar chart for city-wise total number of users
plt.figure(figsize=(10, 6))
plt.bar(username['address'], username['user_id'], color='skyblue')
plt.xlabel('address')
plt.ylabel('Total users ')
plt.title('address-wise Total Number of users (Unique)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

# **top_10_most_frequent_users**

```python
# Grouping users by user and counting the number of users for each username
username = users.groupby('user_id').size().reset_index(name='user_count')

# Merging with username DataFrame to get user names
username = pd.merge(username, users[['user_id', 'username']], on='user_id', how='left')

# Sorting users by username count in descending order
most_frequent_users = username.sort_values(by='user_count', ascending=False)

# Displaying the top 10 most frequent users
top_10_most_frequent_users = most_frequent_users.head(10)
print(top_10_most_frequent_users)
```

## **flights_count_by_airline**

```python
# Grouping flights by airline and counting the number of flight in each airline
flights_count_by_airline = flights.groupby('airline')['flight_id'].count().reset_index()

# Renaming columns for clarity
flights_count_by_airline.columns = ['airline', 'Total flights']

# Sorting airline by total flights in descending order
flights_count_by_airline = flights_count_by_airline.sort_values(by='Total flights', ascending=False)

# Displaying the total number of flights available by airline
print(flights_count_by_airline)
```

# **Distribution of hotels Across room price**

```python
import matplotlib.pyplot as plt

# Grouping hotels by both name and room_price and counting the number of hotels in each room_price
hotels_distribution = hotels.groupby(['name', 'room_price'])['hotel_id'].count().reset_index()

# Plotting the distribution of hotels across room_price
plt.figure(figsize=(12, 8))
for name in hotels_distribution['name'].unique():
    data = hotels_distribution[hotels_distribution['name'] == name]
    plt.bar(data['room_price'], data['hotel_id'], label=name)

plt.xlabel('room_price')
plt.ylabel('Number of hotels')
plt.title('Distribution of hotels Across room_price')
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()
plt.show()
```

# **Month-wise Total bookings**

```python
import pandas as pd
import matplotlib.pyplot as plt

# Convert booking_date to datetime if it's not already in datetime format
bookings['booking_date'] = pd.to_datetime(bookings['booking_date'])

# Extract the month from the booking date
bookings['month'] = bookings['booking_date'].dt.month

# Group the bookings by month and calculate total price for each month
monthly_sales = bookings.groupby('month')['total_price'].sum().reset_index()

# Plotting month-wise total price
plt.figure(figsize=(10, 6))
plt.plot(monthly_sales['month'], monthly_sales['total_price'], marker='o', color='skyblue')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.title('Month-wise Total bookings')
plt.xticks(range(1, 13))
plt.grid(True)
plt.tight_layout()
plt.show()
```
